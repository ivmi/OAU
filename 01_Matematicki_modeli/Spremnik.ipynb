```python
# priprema okruženja
%matplotlib nbagg
from utils import f_ode, resp_plot
import math
```

```python
# Matematički model
def d_spremnik(qu, As, Av):
    g = 9.81
    
    def deriv(t, u, h):                    
        qu=u
        
        if h<0:
            h=0
        
        qi = Av*math.sqrt(2*g*h)
        dh = 1/As * (qu-qi)
        
        return [dh]
    
    return deriv
```

```python
# Odziv
t, y = f_ode(d_spremnik(qu=0.1, As=1, Av=0.01), #matematicki model
            [1],# pocetni uvjeti
             0.01, # ulazni protok
             60)  # vrijeme simulacije

resp_plot(t, y, 'Odziv', ['Razina u spremniku']).show()
```
