```python
# priprema okruženja
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from control import tf, step_response, forced_response, initial_response
import ipywidgets as ipw
```

```python
G = tf([1], [1, 0])  # G=1/s
G
```

```python
t = np.linspace(0, 5, 100)

u = np.hstack((
    np.zeros(20),
    np.ones(20),
    np.zeros(20),
    -2*np.ones(20),
    np.zeros(20),    
),    
)


plt.figure()
plt.step(t,u)
plt.legend(['Ulazni signal'])
plt.grid()
```

```python
ty,y,x = forced_response(G, t, u)

plt.figure()
plt.plot(t, u, ty, y)
plt.legend(['Ulazni signal', 'Izlazni signal'])
#plt.grid()
```
