```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider
from IPython.display import display
from traitlets import observe
```

```python
# sinusni signal

amp_slider = FloatSlider(description='amplituda', min=1, max=10, step=0.1)
freq_slider = FloatSlider(description='frekvencija', min=1, max=10, step=0.1)
phase_slider = FloatSlider(description='fazni pomak', min=0, max=4*np.pi, step=0.1)

def get_data():
    magnitude = amp_slider.value
    freq = freq_slider.value
    phase = phase_slider.value

    t = np.arange(0, 10, 0.05)
    y = magnitude*np.sin(freq*t+phase)
    return t,y

def update_plot(value):
    t, y = get_data()
    l.set_data(t,y)

fig = plt.figure()
fig.set_label('Sinusni signal')

t,y = get_data()
l, = plt.plot(t,y)
plt.grid()
plt.ylim(-10, 10)

amp_slider.observe(update_plot, names='value')    
freq_slider.observe(update_plot, names='value')    
phase_slider.observe(update_plot, names='value')    

update_plot(None)
display(amp_slider, freq_slider, phase_slider)    
```
