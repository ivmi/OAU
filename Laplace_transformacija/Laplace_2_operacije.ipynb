```python
# priprema okruženja
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from control import tf, step_response, forced_response
```

```python
# definiramo 2 prijenosne funkcije

G1 = tf([1], [2, 1])
G2 = tf([2], [10, 1])
print(G1)
print(G2)
```

```python
# odziv prijenosnih funkcija na skokovitu pobudu

t = np.linspace(0, 70, 200)
t1, u1 = step_response(G1, t)
t2, u2 = step_response(G2, t)

plt.figure()
plt.plot(t1, u1, t2, u2)
plt.legend(['G1', 'G2'])
plt.grid()
plt.show()
```

```python
# ako su prijenosne funkcije spojene u seriju (izlaz jednog bloka je ulaz drugog), množimo funkcije

t, u = step_response(G1*G2)

plt.figure()
plt.plot(t, u)
plt.legend(['G1*G2'])
plt.grid()
plt.show()
print(G1*G2)
```

```python
# ako su prijenosne funkcije spojene paralelno, zbrajamo funkcije

t, u = step_response(G1+G2)

plt.figure()
plt.plot(t, u)
plt.legend(['G1+G2'])
plt.grid()
plt.show()
print(G1*G2)
```
